```python
ratings=final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

# Observation

1. When ratings are from 4.5 to 4.9 ----> Excellent
2. When ratings are from 4.0 to 4.0	----> Very Good
3. When ratings are from 3.5 to 3.9 ----> Good
4. When ratings are from 2.5 to 3.4 ----> Average
5. When ratings are from 1.8 to 2.4 ----> Poor

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating", y='Rating Count', data=ratings, hue="Rating color", palette=['gray', 'red','orange','yellow', 'green','darkgreen'])
plt.show()
```

# Observations:
1. Not rated(0 rating)  count is very high 
2. Max ratings are between 2.5 to 3.4

```python
# count plot
sns.countplot(x="Rating color", data=ratings, palette=['gray', 'red','orange','yellow', 'green','darkgreen'])
```
